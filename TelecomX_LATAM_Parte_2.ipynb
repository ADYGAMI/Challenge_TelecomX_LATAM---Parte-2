```python
import pandas as pd

df = pd.read_csv('datos_tratados.csv')
display(df.head())
```

```python
df = df.drop('customerID', axis=1)
display(df.head())
```

```python
categoricas = ['gender','InternetService','Contract','PaymentMethod']

pd.get_dummies(data=df,columns=categoricas, dtype=int).head()
```

```python
datos_codificados =pd.get_dummies(data=df,columns=categoricas, dtype=int)
datos_codificados.sample(5)
```

```python
datos_codificados = datos_codificados.drop('InternetService_No', axis=1)
display(datos_codificados.head())
```

```python
churn_percentage = datos_codificados['Churn'].value_counts(normalize=True) * 100
display(churn_percentage)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
churn_percentage.plot(kind='bar')
plt.title('Percentage of Churn')
plt.xlabel('Churn')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```
